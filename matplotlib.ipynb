**MATPLOTLIB**

```python
import matplotlib.pyplot as plt
```

**linear graph**

```python
x=[1,2,3,4]
y=[2,4,6,8]
c=['r','b','g','y']
plt.plot(x,y)
plt.show()
```

```python
#bar graph
plt.bar(x,y)
plt.show()
```

```python
#coloring the bar graph
plt.bar(x,y,color=c)
plt.show()
```

**BAR PLOT**

```python
x=['python','c','c++','java']
y=[87,89,70,90]
plt.bar(x,y,width=0.4,color='r',align='center',edgecolor='b',linewidth=4,alpha=0.3,label='skills')
plt.legend()
plt.title('skills',fontsize=40)
plt.xlabel('language')
plt.ylabel('popularity')
plt.show()
```

```python
import numpy as np

x=['python','c','c++','java']
y=[87,89,70,90]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.title('skills',fontsize=40)
plt.xlabel('language')
plt.ylabel('popularity')

plt.bar(p,y,width,label='skills')
plt.bar(p1,z,width,label='skills1')
plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
#horizontal bargraph
x=['python','c','c++','java']
y=[87,89,70,90]
plt.barh(x,y,color='r',align='center',linewidth=4,label='skills')
plt.legend()
plt.title('skills',fontsize=40)
plt.xlabel('language')
plt.ylabel('popularity')
plt.show()
```

**SCATTER PLOT**

```python
x=[1,2,3,4,5,6,7]
y=[2,3,4,5,6,6,7]
plt.scatter(x,y,color='r')
plt.title("Scatter plot",fontsize=10)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[2,3,4,5,6,6,7]
plt.scatter(x,y,color='r',marker="*",edgecolor="b")
plt.title("Scatter plot",fontsize=20)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[2,3,4,5,6,6,7]
colors=[20,30,40,50,60,70,80]
plt.scatter(x,y,c=colors,cmap='viridis')
t=plt.colorbar()
t.set_label("Color Bar",fontsize=18)
plt.title("Scatter plot",fontsize=20)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

**STACK PLOT**

```python
x=[1,2,3,4,5]
area=[2,3,2,5,4]
plt.stackplot(x,area)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,7]
area3=[8,7,6,5,4]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','g','y'])
plt.legend(loc=2)
plt.show()
```

```python
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,7]
area3=[8,7,6,5,4]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=['r','g','y'],baseline="wiggle")
plt.legend(loc=2)
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
